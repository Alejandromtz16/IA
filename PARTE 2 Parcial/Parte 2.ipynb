## Librerias 


```python
import numpy as np
from scipy.stats import norm 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt{}

%matplotlib inline
plt.rcParams['figure.figsize'] = (20, 10)
plt.style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')
```

## carga del dataset 

```python
Axis_data = sns.load_dataset("penguins")
Axis_data
```

```python
Axis_data.describe()
```

```python
axis_mu = Axis_data["body_mass_g"].mean() 
axis_sig = Axis_data["body_mass_g"].std() 
x1 =  3550
x2 = 4750
```

```python
z1 = ( x1 - axis_mu ) / axis_sig
z2 = ( x2 - axis_mu ) / axis_sig

x = np.arange(z1, z2, 0.001) 
x_all = np.arange(-10, 10, 0.001) 
y = norm.pdf(x,0,1)
y2 = norm.pdf(x_all,0,1)
```

## Funcion de graficacion 

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.style.use('fivethirtyeight')
ax.plot(x_all,y2)

ax.fill_between(x,y,0, alpha=0.3, color='b')
ax.fill_between(x_all,y2,0, alpha=0.1)
ax.set_xlim([-4,4])
ax.set_xlabel('Número de Desviaciones Estándard Fuera de la Media')
ax.set_yticklabels([])
ax.set_title('Distribución Normal')

plt.show()
```
